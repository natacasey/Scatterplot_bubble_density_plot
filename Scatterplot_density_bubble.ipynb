```python
#importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#reading data to a dataframe
data = pd.read_csv("crimerates-by-state-2005.csv")
data.head()
```

```python
data = data[data['state']!='District of Columbia']
data = data[data['state']!='United States']
burglary = data['burglary']
murder = data['murder']
population = data['population']
states=data['state']
```

```python
#creating a scatterplot
import seaborn as sns
fig, ax = plt.subplots(figsize = (8,6))
sns.regplot(murder, burglary, lowess=True, color="orange",  line_kws={"color":"gray", "linewidth":2})
ax.tick_params(axis=u'both', which=u'both',length=0)
plt.margins(0)
xticks, xticklabels = plt.xticks()
xmin = (3*xticks[0] - xticks[1])/2.
xmax = (3*xticks[-1] - xticks[-2])/2.
plt.xlim(xmin, xmax)
plt.xticks(xticks)
yticks, yticklabels = plt.yticks()
ymin = (2*yticks[0] - yticks[1])/2.
ymax = (2*yticks[-1] - yticks[-2])/2.
plt.ylim(ymin, ymax)
plt.yticks(yticks)
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 1300, 200))
plt.xticks(ha='left',fontsize=12, color = 'dimgray')
plt.yticks(fontsize=12, color = 'dimgray')
right_side = ax.spines["right"]
top_side = ax.spines["top"]
right_side.set_visible(False)
top_side.set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.set_title('\nMURDERS VERSUS BURGLARIES IN THE US\n', fontsize=17,  loc='left', color='dimgray')
ax.set_ylabel('\nBurglaries (per 100,000 population)', fontsize=14, labelpad=12, color='gray')
ax.set_xlabel('\nMurders (per 100,000 population)', fontsize=14, labelpad=1, color='gray')
ax.tick_params(axis='x', labelsize=13) 
ax.tick_params(axis='y', labelsize=13) 
plt.show();
```

```python
#creating a density plot
import seaborn as sns
fig, ax = plt.subplots(figsize = (8,6))
sns.kdeplot(burglary,color = 'orange', shade=True)
ax.get_legend().remove()
plt.xticks(ha='left',fontsize=12, color = 'dimgray')
plt.yticks(fontsize=12, color = 'dimgray')
ax.tick_params(axis=u'both', which=u'both',length=0)
plt.margins(0)
right_side = ax.spines["right"]
top_side = ax.spines["top"]
right_side.set_visible(False)
top_side.set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.set_title('\nDISTRIBUTION OF BURGLARIES IN THE US\n', fontsize=17,  loc='left', color='dimgray')
ax.set_ylabel('\nDensity', fontsize=14, labelpad=12, color='gray')
ax.set_xlabel('\nBurglaries (per 100,000 population)', fontsize=14, labelpad=1, color='gray')
ax.tick_params(axis='x', labelsize=13) 
ax.tick_params(axis='y', labelsize=13) 
plt.xlim([0, 1500])
plt.ylim([0, 0.0015])
plt.show();
```

```python
#creating a bubble chart
fig, ax = plt.subplots(figsize = (16,12))
sc = plt.scatter(x =data['murder'], y=data['burglary'], s=population * 0.00016,alpha=0.6,color = 'orange', linewidth=2)
ax.tick_params(axis=u'both', which=u'both',length=0)
plt.margins(0)
xticks, xticklabels = plt.xticks()
xmin = (3*xticks[0] - xticks[1])/2.
xmax = (3*xticks[-1] - xticks[-2])/2.
plt.xlim(xmin, xmax)
plt.xticks(xticks)
yticks, yticklabels = plt.yticks()
# shift half a step to the left
# x0 - (x1 - x0) / 2 = (3 * x0 - x1) / 2
ymin = (2*yticks[0] - yticks[1])/2.
# shaft half a step to the right
ymax = (2*yticks[-1] - yticks[-2])/2.
plt.ylim(ymin, ymax)
plt.yticks(yticks)
plt.xticks(np.arange(0, 11, 1))
plt.yticks(np.arange(0, 1300, 100))
for line in range(0,data.shape[0]):
    if line ==26 or line == 47 or line == 44 or  line ==22 or line == 10 or line ==33:
        
        plt.text(data.murder.iloc[line], data.burglary.iloc[line], data.state.iloc[line], horizontalalignment='right',  size=8, color='black', weight='normal', wrap=True)
    elif line ==16:
        plt.text(data.murder.iloc[line], data.burglary.iloc[line], data.state.iloc[line], horizontalalignment='center',  size=8, color='black', weight='normal', wrap=True)
    else:
        plt.text(data.murder.iloc[line], data.burglary.iloc[line], data.state.iloc[line], horizontalalignment='left',  size=8, color='black', weight='normal', wrap=True)
plt.xticks(ha='left',fontsize=12, color = 'dimgray')
plt.yticks(fontsize=12, color = 'dimgray')
ax.tick_params(axis=u'both', which=u'both',length=0)
right_side = ax.spines["right"]
top_side = ax.spines["top"]
right_side.set_visible(False)
top_side.set_visible(False)
ax.spines['left'].set_color('gray')
ax.spines['bottom'].set_color('gray')
ax.set_title('\nMURDERS VERSUS BURGLARIES IN THE US SIZED BY POPULATION\n', fontsize=17,  loc='left', color='dimgray')
ax.set_ylabel('\nBurglaries (per 100,000 population)', fontsize=14, labelpad=12, color='gray')
ax.set_xlabel('\nMurders (per 100,000 population)', fontsize=14, labelpad=1, color='gray')
ax.tick_params(axis='x', labelsize=13) 
ax.tick_params(axis='y', labelsize=13) 
plt.xlim([0, 11])
# produce a legend with a cross section of sizes from the scatter
#handles, labels = sc.legend_elements(prop="sizes", alpha=0.6)
kw = dict(prop="sizes", num=3, alpha = 0.4, color='orange')
legend2 = ax.legend(*sc.legend_elements(**kw),
                    loc="upper right", title="Population (K)",frameon=False, labelspacing=2, borderaxespad=0.)
ax.add_artist(legend2)
#ax.legend(frameon=False, labelspacing=3, borderaxespad=0., bbox_to_anchor=(0.53, 0.53, 0.53, 0.53),title="Population in thousands")
plt.savefig('bubble.png')
plt.show();
```
